```python
import torch
import pandas as pd
import numpy as np

train = pd.read_csv('train.csv')
train.columns
```

```python
train.head(10)
```

For Version 1,
- Remove the following columns: Name, Ticket, Cabin
- Split Embarked into Embarked_C, Embarked_S and Embarked_Q
- Normalize Fare and Age

```python
embarked_c = [1 if x == 'C' else 0 for x in train['Embarked']]
embarked_q = [1 if x == 'Q' else 0 for x in train['Embarked']]
embarked_s = [1 if x == 'S' else 0 for x in train['Embarked']]
train['Embarked_S'] = embarked_s
train['Embarked_C'] = embarked_c
train['Embarked_Q'] = embarked_q
sex = [1 if x=='male' else 0 for x in train['Sex']]
train['Sex'] = sex
mean_age = train['Age'].mean()
std_age  = train['Age'].std()
train['Age'] = (train['Age'] - mean_age)/std_age
mean_fare = train['Fare'].mean()
std_fare  = train['Fare'].std()
train['Fare'] = (train['Fare'] - mean_fare)/std_fare
train = train.drop(columns=['Name', 'Ticket', 'Cabin', 'Embarked'])
train = train.fillna(0)
train.head(10)
```
